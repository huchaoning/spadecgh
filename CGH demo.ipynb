```python
# 导入库 (cgh 库计划有对应的 cli 版本, 但是目前还没写. 直接从 python 使用没有问题)
from hduq.cgh import *
```

```python
# 实例化 CGH
cgh = CGH(103) # 103um 表示特征宽度
```

```python
# 往该实例中添加模式, 模块采用惰性计算, 先添加要分解的模式再进行计算
cgh.add_modes(HG(0, 1), 500, 0) # 500, 0 分别为 x 和 y 方向的载波频率, 是一个相对值
```

```python
# 生成的 cgh 预览 (未执行计算会自动调用 cal 方法计算)
cgh.show()
```

```python
# 正负模式的生成方法
P = HG(0, 0) + HG(0, 1) # 正模式
M = HG(0, 0) - HG(0, 1) # 副模式
```

```python
# 可以产生很复杂的叠加模式 (归一化系数自动为等权重叠加)
_ = HG(0, 0) + HG(0, 1) + HG(0, 2) + HG(0, 3)
```

```python
# 同时模式分解
cgh = CGH(103) # 替换掉原来 CGH 实例, 因为现在要产生一副新的 CGH 了
```

```python
# 传入要分解的模式
cgh.add_modes([P, M], [500, 500], [-50, 50]) # 一次性传入

# 也可以分两次传入
# cgh.add_modes(P, 500, -50)
# cgh.add_modes(M, 500, 50)
```

```python
# 直接读取结果或者保存
cgh_array = cgh.result()
# cgh.save('your_cgh.bmp')
```
